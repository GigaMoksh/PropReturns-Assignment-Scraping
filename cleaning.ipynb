```python
import pandas as pd
from googletrans import Translator
```

```python
translator = Translator()
```

```python
df = pd.read_csv('data.csv', delimiter=':')
```

```python
df
```

```python
df['property_desc'][1][:3]
```

```python
df['property_desc'] = df['property_desc'].apply(lambda x: x[2:])
```

```python
df['property_desc']
```

```python
df.head(5)
```

```python
df['rdate'] = pd.to_datetime(df['rdate'])
```

```python
df.head(5)
```

```python
df['seller_name'] = df['seller_name'].apply(lambda x: x[1:-1])
```

```python
df['seller_name']
```

```python
df.head(5)
```

```python
import numpy as np
df = df.replace(r'^\s*$', np.nan, regex=True)
```

```python
df.head(5)
```

```python
df.to_csv('cleaned_data.csv', index=False, sep=":")
```

```python
for col in df.columns:
    if col not in ['doc_no', 'rdate', 'sro_code', 'status']:
        df[col] = df[col].apply(lambda x: translator.translate(x, src='mr', dest='en').text)
```
